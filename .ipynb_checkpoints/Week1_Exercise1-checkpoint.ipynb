```python
def haversine(lat1, lon1, lat2, lon2):
    import numpy as np
    R=6378.137
    lat1, lon1, lat2, lon2 = map(np.deg2rad, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    total_m = (R * c)*1000
    return total_m
```

```python
dist=haversine(dat['lat'][1:],dat['lon'][1:],dat['lat'][:-1],dat['lon'][:-1])
dist_order=dist.cumsum()
```

Geostrophic vertical shear using the thermal wind equation The relationship between the
geostrophic vertical shear and the horizontal density gradient is given by the thermal wind equation
-fdv/dz=(g/rho0)(drhodx)

```python
#compute coriolis acceleration
def coriolis_acc(lat,omega=0.729*10**-4):
    return 2*omega*np.sin(lat/360.*2*np.pi)

f=coriolis_acc(dat['lat'])
g=9.8 # define gravitaional acceleration
ref_dens=1027.
geo_vert_shear=((g/ref_dens)*(del_pot_dens0))/f[:-1]
#plot
cf=plt.contourf(xx/1000,yy,geo_vert_shear.T)
plt.gca().invert_yaxis()
plt.xlabel('Distance from {} to {} (km)'.format(dat['lat'][0],dat['lat'][-1]))
plt.ylabel('Depth (m)')
plt.title('Geostrophic Vertical Shear')
cbar=plt.colorbar(cf)
cbar.set_label('s$^{-1}$')
```

```python
#### Compute geostrophic velocity with level of no motion at 1000m
idx=abs(dat['pres_en_cours']-1000.).argmin()
vel_no_motion=np.reshape(np.repeat(geo_vert_shear[:,idx],2500),(62,2500))
vel_no_motion_bot=np.reshape(np.repeat(geo_vert_shear[:,-1],2500),(62,2500))
vel_no_motion_surf=np.reshape(np.repeat(geo_vert_shear[:,0],2500),(62,2500))

geo_vel1000=np.sum(geo_vert_shear-vel_no_motion,axis=1)
geo_vel_bot=np.sum(geo_vert_shear-vel_no_motion_bot,axis=1)
geo_vel_surf=np.sum(geo_vert_shear-vel_no_motion_bot,axis=1)

# geo_vel0=np.sum(geo_vert_shear,axis=0)

plt.plot(geo_vel1000)
plt.plot(geo_vel_bot)
plt.plot(geo_vel_surf)


# plt.plot(geo_vel1000,dat['pres_en_cours'][:,0:idx].flatten(),label='no_motion=1000m')
# plt.plot(geo_vel0,dat['pres_en_cours'].flatten(),label='no_motion=0')
# plt.legend()
# plt.gca().invert_yaxis()
# plt.xlabel('Geostrophic Velocity (m.s$^{-2}$)')
# plt.ylabel('depth')
# plt.title('Geostrophic velocity across the full section')
```
