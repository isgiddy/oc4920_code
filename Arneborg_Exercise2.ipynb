# Turbulence Exercise 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plot the Universal velocity spectrum

```python
v=10**-6    #### kinematic viscocity
e=10**-10   ###dissipation rate
```

```python
def theta(G2,e):
    return (e**(3/4))*(v**(-1/4))*G2
```

```python
G2=np.asarray([0.5285,0.6397,0.7763,0.9404,1.138,1.380,1.682,2.302,2.176,1.373,0.5278,0.1342,0.0616,0.0249,0.0101])
k=np.asarray([2.83*10**-4,5.03*10**-4,8.95*10**-4,1.59*10**-3,2.83*10**-3,5.03*10**-3,8.95*10**-3,1.59*10**-2,2.83*10**-2,5.03*10**-2,7.97*10**-2,1.26*10**-1,1.59*10**-1,2.*10**-1,2.52*10**-1])
```

```python
func1=theta(G2,e)
func2=theta(G2,e=10**-9)
func3=theta(G2,e=10**-8)
func4=theta(G2,e=10**-7)
```

```python
plt.loglog(k,func1,label='e=10$^{-10}$')
plt.loglog(k,func2,label='e=10$^{-9}$')
plt.loglog(k,func3,label='e=10$^{-8}$')
plt.loglog(k,func4,label='e=10$^{-7}$')
plt.xlabel('k/ks [1/m]')
plt.ylabel('Velocity Shear Spectrum [s$^{-2}$ cpm$^{-1}$ ]')
plt.legend(loc=3)
```

The kolmogorov scale is where there is a peak in the spectrum 

## Compute dissipation from shear variance

```python
### Take the integral
shear_variance=np.sum(func1[:-1]*np.diff(k))

# multiply by 7.5 and the kinematic viscocity
E=7.5*v*shear_variance


print('shear_variance,e={}: {}'.format(10**-10,shear_variance))
print('disspation rate,e={}: {}'.format(10**-10,E))

shear_variance2=np.sum(func2[:-1]*np.diff(k))
E2=7.5*v*shear_variance2
print('shear_variance,e={}: {}'.format(10**-9,shear_variance2))
print('disspation rate,e={}: {}'.format(10**-9,E2))

shear_variance3=np.sum(func3[:-1]*np.diff(k))
E3=7.5*v*shear_variance3
print('shear_variance,e={}: {}'.format(10**-8,shear_variance3))
print('disspation rate,e={}: {}'.format(10**-8,E3))

shear_variance4=np.sum(func4[:-1]*np.diff(k))
E4=7.5*v*shear_variance4

print('shear_variance,e={}: {}'.format(10**-7,shear_variance4))
print('disspation rate,e={}: {}'.format(10**-7,E4))
```

Shear variance can be calculated from the integral of the shear spectrum. Using this method, I calculate dissipation rates smaller than the disspation rate used to produce the curve.
